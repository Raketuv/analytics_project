```python
## pandas

import pandas as pd

## random forest tree

from sklearn.ensemble import RandomForestRegressor

from sklearn import metrics

## OLS

import statsmodels.api as sm

from statsmodels.formula.api import ols
```

```python
# Read in data

df = pd.read_csv("train.csv")
df2 = df
```

```python
# Create Total Columns as sum of individual columns for bathrooms no and porch size

df2["TotalBathrooms"] = df.BsmtFullBath + df.FullBath + 0.5*df.BsmtHalfBath + 0.5*df.HalfBath
df2["TotalPorchSize"] = df.WoodDeckSF+df.OpenPorchSF+df.EnclosedPorch+df["3SsnPorch"]+df.ScreenPorch
```

```python
# Replace values  
#       9	Excellent
#       8	Very Good
#       7	Good
#       6	Above Average
#       5	Average
#       4	Below Average
#       3	Fair
#       2	Poor

#       Ex	Excellent
#       Gd	Good
#       TA	Average/Typical
#       Fa	Fair
#       Po	Poor

df2 = df2.replace(["TA","Gd","Ex","Fa","Po"], [5,7,9,3,2])

```

```python
# MSSubClass is actually categorical:

#        20	1-STORY 1946 & NEWER ALL STYLES
#        30	1-STORY 1945 & OLDER
#        40	1-STORY W/FINISHED ATTIC ALL AGES
#        45	1-1/2 STORY - UNFINISHED ALL AGES
# etc

df2["MSSubClass"] = df2["MSSubClass"].astype(str)
```

```python
# Convert other categorical values into dummies

df2 = pd.get_dummies(df2)
```

```python
# Remove na values

df2.isna().sum().sort_values().tail(10)
```

```python
# Check out PoolQC

df2["PoolQC"].value_counts()
```

```python
# Check out FireplaceQu

df2["FireplaceQu"].value_counts()
```

```python
# Check out LotFrontage

df2["LotFrontage"].value_counts()
```

```python
# Drop PoolQC, FireplaceQu

df2 = df2.drop(columns=["PoolQC","FireplaceQu"])
```

```python
# Check remaining columns with missing values

df2.isna().sum().sort_values().tail(10)
```

```python
# Make into list

list(df2.isna().sum().sort_values().tail(7).index)
```

```python
# Fill missing values with median 

for i in list(df2.isna().sum().sort_values().tail(7).index):
    df2[i] = df2[i].fillna(df2[i].median())
```

```python
# Run random forest regression with SalePrice as dependent variable

m = RandomForestRegressor(n_jobs=-1)
m.fit(df2.drop('SalePrice', axis=1), df2.SalePrice)
```

```python
# Create dataframe with feature importance from random forest model

featureimportance = pd.DataFrame()
featureimportance["feature"] = df2.drop(columns="SalePrice").columns
featureimportance["importance"] = m.feature_importances_
featureimportance.sort_values(by='importance', ascending=False)
```

```python
# Cut off all features with importance under .005

featureimportance[featureimportance["importance"]>0.005].sort_values(by='importance',ascending=False)
```

```python
# Make list of important cols

importantcols = list(featureimportance[featureimportance["importance"]>0.005]["feature"])
```

```python
# Add Sale Price lol

importantcols += ["SalePrice"]
importantcols
```

```python
# Correlation heat map for important columns

df2[importantcols].corr().style.background_gradient(cmap='coolwarm')
```

```python
# Drop columns: Year built, since YearRemodAdd is better, BsmtFinSF1 since we rather keep TotalBsmtSF, 
# first and second floor since we rather have total "GrLivArea".
# Drop Total rooms above ground since GrLivArea contains that
# Drop Garage area since garage cars is better.
# Garage year built since its not nice

df3 = df2[importantcols].drop(columns=["YearBuilt","BsmtFinSF1","1stFlrSF","2ndFlrSF","TotRmsAbvGrd","GarageArea","GarageYrBlt"])
```

```python
# Again, heatmap

df3.corr().style.background_gradient(cmap='coolwarm')
```

```python
# Do OLS with remaining columns

X = df3.drop(columns="SalePrice")
X = sm.add_constant(X) # We must add the intercept using the add_constant function
Y = df3.SalePrice

sm.OLS(Y, X).fit().summary()
```

```python
# Lot frontage seems just like a different dimension of lotarea. do OLS without that.

X = df3.drop(columns="SalePrice").drop(columns="LotFrontage")
X = sm.add_constant(X) # We must add the intercept using the add_constant function
Y = df3.SalePrice

sm.OLS(Y, X).fit().summary()
```
